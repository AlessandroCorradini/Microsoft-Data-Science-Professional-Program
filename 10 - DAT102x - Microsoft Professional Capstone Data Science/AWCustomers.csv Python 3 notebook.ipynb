```python
import pandas as pd
frame['BirthDate'] =  pd.to_datetime(frame['BirthDate'])

frame.dtypes
```

```python
def convert_age(df):
    from datetime import date
    import pandas as pd
    today = date.today()
    age = []
    for bday in df["BirthDate"]:
        age.append(2017 - int(bday.year))
    df["age"] = age
    return df

convert_age(frame)
```
